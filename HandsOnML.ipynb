```python
# Define a quadratic function
def quadratic_function(x):
    return x**2 - 4*x + 3  # i.e f(x) = x^2 - 4x + 3

# Generate x values
x = np.linspace(-2, 6, 100)  # i.e 100 points from -2 to 6

# Calculate the corresponding y values
y = quadratic_function(x)

# Create the plot
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='f(x) = x^2 - 4x + 3', color='green')
plt.title('Quadratic Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.axhline(0, color='black',linewidth=0.5, ls='--')  # x-axis
plt.axvline(0, color='black',linewidth=0.5, ls='--')  # y-axis
plt.grid()
plt.legend()
plt.show()
```
