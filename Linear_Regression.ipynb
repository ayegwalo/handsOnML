```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# URL to access the raw file from GitHub
data_url = "https://raw.githubusercontent.com/ageron/data/main/lifesat/lifesat.csv"

# Load the life satisfaction dataset
lifesat = pd.read_csv(data_url)

# Extract the 'GDP per capita (USD)' column as a NumPy array for feature variable
x = lifesat[["GDP per capita (USD)"]].values

# Extract the 'Life satisfaction' column as a NumPy array for target variable
y = lifesat[["Life satisfaction"]].values
```

```python
# Create a scatter plot to visualize the relationship between GDP per capita and life satisfaction
lifesat.plot(kind='scatter', grid=True,
             x="GDP per capita (USD)", y="Life satisfaction")

# Set the limits for the x-axis (GDP per capita) and y-axis (Life satisfaction)
plt.axis([23_500, 62_500, 4, 9])

# Get the predicted values for the regression line
y_pred = model.predict(x)

# Plot the regression line
plt.plot(lifesat["GDP per capita (USD)"], y_pred, color='red', linewidth=3, label="Regression Line")

# Display and show
plt.show()
```

```python
# An instance of LinearRegression model
model = LinearRegression()

# Fit the model to the training data (x and y)
model.fit(x, y)

# Define a new input value for prediction (Prediction)
x_new = [[37_655.2]] 

# Print the predicted output for the new input value
print(model.predict(x_new))
```
